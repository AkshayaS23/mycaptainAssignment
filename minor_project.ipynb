```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Dataset URL
url = "https://github.com/AkshayaS23/mycaptainAssignment/blob/main/bankdetails.csv"
df = pd.read_csv(url)
# Check for missing values
missing_values = df.isnull().sum()

if missing_values.any():
    df = df.dropna()

print("After dropping missing values:")
print(df.isnull().sum())

car_min_price = df[df['price'] == df['price'].min()]
print("Car(s) with minimum price:")
print(car_min_price[['company', 'price']])


car_max_price = df[df['price'] == df['price'].max()]
print("\nCar(s) with maximum price:")
print(car_max_price[['company', 'price']])

body_style_counts = df['body-style'].value_counts()
print("\nNumber of cars by body style:")
print(body_style_counts)

avg_price = df.groupby('body-style')['price'].mean().sort_values(ascending=False)
print("\nAverage price of cars by body style (descending order):")
print(avg_price)

cylinder_counts = df['num-of-cylinders'].value_counts()
print("\nCount of car types by number of cylinders:")
print(cylinder_counts)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_price.index, y=avg_price.values, hue=avg_price.index, palette='viridis', dodge=False, legend=False)
plt.xticks(rotation=45)
plt.title('Average Price of Cars by Body Style')
plt.xlabel('Body Style')
plt.ylabel('Average Price')
plt.tight_layout()
plt.show()

import matplotlib.pyplot as plt
import seaborn as sns
avg_price_by_company = df.groupby('company')['price'].mean().sort_values(ascending=False)
avg_price_by_body_type = df.groupby('body-style')['price'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
sns.barplot(x=avg_price_by_company.index, y=avg_price_by_company.values)
plt.title('Average Price by Company')
plt.xticks(rotation=90)

plt.subplot(2, 2, 2)
sns.barplot(x=avg_price_by_body_type.index, y=avg_price_by_body_type.values)
plt.title('Average Price by Body Type')
plt.xticks(rotation=90)

plt.subplot(2, 2, 3)
sns.countplot(data=df, x='body-style', hue='num-of-cylinders')
plt.title('Body Type & Number of Cylinders')
plt.xticks(rotation=90)

plt.subplot(2, 2, 4)
sns.scatterplot(data=df, x='horsepower', y='engine-type')
plt.title('Engine Type & Horse Power')

plt.tight_layout()
plt.show()








```
